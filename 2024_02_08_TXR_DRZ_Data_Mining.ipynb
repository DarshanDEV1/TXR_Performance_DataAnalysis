```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('TXR_OVR_Report.csv')
```

```python
df.head()
```

```python
df.describe().head()
```

```python
def _plot_series(xaxis, yaxis, series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series[xaxis]
  ys = series[yaxis]

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

#fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Time Stamp', ascending=True)
```

```python
_plot_series('Time Stamp', 'cpu_level', df_sorted, '')
#sns.despine(fig=fig, ax=ax)
plt.xlabel('Time Stamp')
_ = plt.ylabel('cpu_level')
```

```python
_plot_series('Time Stamp', 'gpu_level', df_sorted, '')
#sns.despine(fig=fig, ax=ax)
plt.xlabel('Time Stamp')
_ = plt.ylabel('gpu_level')
```

```python
_plot_series('Time Stamp', 'average_frame_rate', df_sorted, '')
#sns.despine(fig=fig, ax=ax)
plt.xlabel('Time Stamp')
_ = plt.ylabel('average_frame_rate')
```

```python
_plot_series('Time Stamp', 'gpu_frequency_MHz', df_sorted, '')
#sns.despine(fig=fig, ax=ax)
plt.xlabel('Time Stamp')
_ = plt.ylabel('gpu_frequency_MHz')
```

```python
_plot_series('Time Stamp', 'cpu_frequency_MHz', df_sorted, '')
#sns.despine(fig=fig, ax=ax)
plt.xlabel('Time Stamp')
_ = plt.ylabel('cpu_frequency_MHz')
```
